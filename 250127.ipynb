```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time

# 配置 ChromeDriver 的路径
driver_path = r"C:\Users\flySand\Downloads\chromedriver.exe"  # 用一个更简单的路径

# 创建一个 Service 对象来指定驱动路径
service = Service(driver_path)

# 配置 Chrome 浏览器的选项
options = Options()
options.add_argument("--disable-gpu")
options.add_argument("--no-sandbox")
options.add_argument("--headless")  # 可选：启用无头模式

# 设置 Chrome 浏览器的路径（如果需要）
options.binary_location = r"C:\Program Files\Google\Chrome\Application\chrome.exe"  # 确保路径正确

# 启动 Chrome 浏览器
driver = webdriver.Chrome(service=service, options=options)

# 打开目标网站
url = 'https://jt.dushiquan.net/beijing/'
driver.get(url)

# 使用 WebDriverWait 来等待页面的特定元素加载完毕
wait = WebDriverWait(driver, 10)
wait.until(EC.presence_of_element_located((By.XPATH, "//*[@id='lineListContent']")))

# 获取北京地铁简介
intro_element = driver.find_element(By.XPATH, "//*[@id='lineListContent']/div[1]/div[1]")
intro_text = intro_element.text
print(f"北京地铁简介：\n{intro_text}\n")

# 获取北京地铁其他介绍
intro_section_element = driver.find_element(By.XPATH, "//*[@id='lineListContent']/div[1]/div[3]")
intro_section_text = intro_section_element.text
print(f"北京地铁其他介绍：\n{intro_section_text}\n")

# 提取所有线路的详细信息（1号线、2号线...）
lines = driver.find_elements(By.XPATH, "//*[@id='lineListContent']/div")
all_line_data = []

for line in lines:
    # 获取线路标题（例如：北京地铁1号线）
    try:
        line_title = line.find_element(By.XPATH, "div[1]").text
    except:
        continue  # 如果没有找到标题，则跳过

    # 获取线路图与站名（对于每条线路）
    try:
        line_map_and_stations = line.find_element(By.XPATH, "div[2]").text
    except:
        continue  # 如果没有找到站点，则跳过

    # 处理线路信息，将站点名分割并保存
    stations = line_map_and_stations.split('\n')
    line_data = {
        "线路标题": line_title,
        "站点": stations
    }
    all_line_data.append(line_data)

    # 输出每条线路的信息
    print(f"线路标题：{line_title}")
    print(f"站点列表：{stations}")
    print('-' * 50)

# 将数据转换为 Pandas DataFrame（表格格式）
lines_df = pd.DataFrame(all_line_data)

# 展示表格数据
print("地铁线路信息：")
print(lines_df)

# 关闭浏览器
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time

# 配置 ChromeDriver 的路径
driver_path = r"C:\Users\flySand\Downloads\chromedriver.exe"  # 用一个更简单的路径

# 创建一个 Service 对象来指定驱动路径
service = Service(driver_path)

# 配置 Chrome 浏览器的选项
options = Options()
options.add_argument("--disable-gpu")
options.add_argument("--no-sandbox")
options.add_argument("--headless")  # 可选：启用无头模式

# 设置 Chrome 浏览器的路径（如果需要）
options.binary_location = r"C:\Program Files\Google\Chrome\Application\chrome.exe"  # 确保路径正确

# 启动 Chrome 浏览器
driver = webdriver.Chrome(service=service, options=options)

# 打开目标网站
url = 'https://jt.dushiquan.net/beijing/'
driver.get(url)

# 使用 WebDriverWait 来等待页面的特定元素加载完毕
wait = WebDriverWait(driver, 10)
wait.until(EC.presence_of_element_located((By.XPATH, "//*[@id='lineListContent']")))

# 提取所有线路的详细信息（1号线、2号线...）
lines = driver.find_elements(By.XPATH, "//*[@id='lineListContent']/div")
all_line_data = []

# 提取线路标题和对应的站点信息
for line in lines:
    try:
        line_title = line.find_element(By.XPATH, "div[1]").text
    except:
        continue  # 如果没有找到标题，则跳过

    try:
        line_map_and_stations = line.find_element(By.XPATH, "div[2]").text
    except:
        continue  # 如果没有找到站点，则跳过

    # 检查是否为有效的线路信息（避免提取简介等不相关信息）
    if line_title.startswith('北京地铁'):
        stations = line_map_and_stations.split('\n')
        all_line_data.append({
            "线路标题": line_title,
            "站点": stations
        })

# 获取每条线路的站点数据
lines_titles = [data["线路标题"] for data in all_line_data]
stations_lists = [data["站点"] for data in all_line_data]

# 获取最大站点数（确保每列都有相同的行数）
max_stations = max(len(stations) for stations in stations_lists)

# 创建一个空的 DataFrame，初始化列为线路标题
df = pd.DataFrame(columns=lines_titles)

# 将每条线路的站点数据添加到对应的列中
for i, stations in enumerate(stations_lists):
    # 补齐站点数量，使得每列长度一致
    stations += [''] * (max_stations - len(stations))
    df[lines_titles[i]] = stations

# 展示表格数据
print("地铁线路信息：")
print(df)

# 保存为 CSV 文件
df.to_csv('beijing_subway_lines2.csv', index=False, encoding='utf-8')

# 关闭浏览器
driver.quit()

print("CSV 文件已成功生成！")
```
